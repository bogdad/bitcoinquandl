```julia
# trying to get a stddev
std(convert(Array, btcltcdf[:, :Last]))
```

```julia
btcltcta = TimeArray(btcltcdf, colnames=[:High, :Low, :Last], timestamp=:Date)
```

```julia
meanbtcltc=moving(btcltcta, mean, 100)
```

```julia
stdbtcltc=moving(btcltcta, std, 100)
```

```julia
meanbtcltc, meanplusstdbtcltc
```

```julia
meanplusstdbtcltc=stdbtcltc.+meanbtcltc;
meanminusstdbtcltc=-stdbtcltc.+meanbtcltc;
```

```julia
plot()
plot!(btcltcta["Last"], label="BTC LTC rate")
plot!(meanbtcltc["Last"], label="mean")
plot!(meanplusstdbtcltc["Last.+Last"], label="Std + BTC LTC")
plot!(meanminusstdbtcltc["-Last.+Last"], label="Std - BTC LTC")
```
